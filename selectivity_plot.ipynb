```python
import matplotlib.pyplot as plt
import numpy as np

sel=np.load("./sel_resnet20.npy", allow_pickle=True)
imp=np.load("./imp_resnet20.npy", allow_pickle=True)

# plt.scatter(x, y)
# plt.show()
```

```python
plt.title("resnet20 stage:1 16 channel")
plt.xlabel("selectivity index")
plt.ylabel("impact on loss")
plt.scatter(sel[0], imp[0])
plt.show()
```

```python
plt.title("resnet20 stage:2 32 channel")
plt.xlabel("selectivity index")
plt.ylabel("impact on loss")
plt.scatter(sel[1], imp[1])
plt.show()
```

```python
plt.title("resnet20 stage:3 64 channel")
plt.xlabel("selectivity index")
plt.ylabel("impact on loss")
plt.scatter(sel[2], imp[2])
plt.show()
```

```python
plt.title("resnet20 stage:(1,2,3) channel: (16, 32, 64)")
plt.xlabel("selectivity index")
plt.ylabel("impact on loss")
plt.scatter(sel[0], imp[0], label='stage1', color='red')
plt.scatter(sel[1], imp[1], label='stage2', color='green')
plt.scatter(sel[2], imp[2], label='stage3', color='blue')
plt.legend()
plt.show()
```
