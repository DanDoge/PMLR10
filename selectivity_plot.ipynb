```python
import matplotlib.pyplot as plt
import numpy as np

sel=np.load("./sel_resnet20.npy", allow_pickle=True)
imp=np.load("./imp_resnet20.npy", allow_pickle=True)

# plt.scatter(x, y)
# plt.show()
```

```python
plt.title("resnet20 stage:1 16 channel")
plt.xlabel("selectivity index")
plt.ylabel("impact on loss")
plt.scatter(sel[0], imp[0])
plt.show()
```

```python
plt.title("resnet20 stage:2 32 channel")
plt.xlabel("selectivity index")
plt.ylabel("impact on loss")
plt.scatter(sel[1], imp[1])
plt.show()
```

```python
plt.title("resnet20 stage:3 64 channel")
plt.xlabel("selectivity index")
plt.ylabel("impact on loss")
plt.scatter(sel[2], imp[2])
plt.show()
```

```python
plt.title("resnet20 stage:(1,2,3) channel: (16, 32, 64)")
plt.xlabel("selectivity index")
plt.ylabel("impact on loss")
plt.scatter(sel[0], imp[0], label='stage1', color='red', s=10)
plt.scatter(sel[1], imp[1], label='stage2', color='green', s=10)
plt.scatter(sel[2], imp[2], label='stage3', color='blue', s=10)
plt.legend()
plt.show()
```

```python
cor = []
cor.append(np.corrcoef(sel[0], imp[0])[0][1])
cor.append(np.corrcoef(sel[1], imp[1])[0][1])
cor.append(np.corrcoef(sel[2], imp[2])[0][1])
```

```python

l1=plt.plot([1,2,3],cor,'ro-')
plt.title('Resnet20, CIFAR10 ACC: 91.73% ')
plt.xlabel('stage')
plt.ylabel('correlation between impact and selectivity')
plt.legend()
plt.show()
```
